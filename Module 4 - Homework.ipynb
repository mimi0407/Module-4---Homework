```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline 
```

```python
df.columns
```

```python
sns.lmplot(x='PRICE' , y='ZIPCODE' , data=df)
```

```python
corr = stats_df.corr()
sns.heatmap(corr, vmin=-1, annot=True)
```
